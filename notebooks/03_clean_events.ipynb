```python
games = get_games("2024-06-08")
games.head()
```

```python
game_id = games.loc[0, "gameID"]

events = get_game_events(game_id)
throws = clean_game_events(events)
player_stats = calculate_thrower_stats(throws)

player_stats.sort_values("attempts", ascending=False).head(20)
```
